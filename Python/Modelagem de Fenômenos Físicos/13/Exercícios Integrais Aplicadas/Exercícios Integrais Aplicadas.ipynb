```python
import matplotlib.pyplot as pyplot
import numpy as np
import sympy as sp
```

# Exercício 1

```python
x = sp.Symbol("x")
v1 = x**2
v2 = x**2 - 2*x + 1

min_x = 1
max_x = 2
x_lim = 1

# Calcula os valores das funções
vals_x = np.linspace(min_x - x_lim, max_x + x_lim, 500)
vals_v1 = np.array([v1.subs(x, i) for i in vals_x], dtype=float)
vals_v2 = np.array([v2.subs(x, i) for i in vals_x], dtype=float)
vals_x_i = np.linspace(min_x, max_x, 500)
vals_v1_i = np.array([v1.subs(x, i) for i in vals_x_i], dtype=float)
vals_v2_i = np.array([v2.subs(x, i) for i in vals_x_i], dtype=float)

# Gráfico
#pyplot.gca().set_aspect('equal', 'datalim')
pyplot.xlabel(r"$x$")
pyplot.ylabel(r"$y$")
pyplot.axhline(y=0, linestyle="--", color="black")
pyplot.axvline(x=min_x, linestyle="--", color="blue")
pyplot.axvline(x=max_x, linestyle="--", color="blue")

pyplot.plot(vals_x, vals_v1, color="red", label=r"$x^2$")
pyplot.plot(vals_x, vals_v2, color="red", label=r"$x^2 - 2x + 1$")
pyplot.fill_between(vals_x_i, vals_v1_i, y2=vals_v2_i, step="pre", alpha=0.4)

# Mostra resultados
area1 = sp.integrate(v1, (x, min_x, max_x))
area2 = sp.integrate(v2, (x, min_x, max_x))
print(f"Solução: ({np.abs(area1 - area2)})\n")
pyplot.legend()
pyplot.show()
```

# Exercício 2

```python
y = sp.Symbol("y")
x1 = y**2
x2 = sp.sqrt(y)

min_y = 1
max_y = 3
y_lim = 1

# Calcula os valores das funções
vals_y = np.linspace(min_y - y_lim, max_y + y_lim, 500)
vals_x1 = np.array([x1.subs(y, i) for i in vals_y], dtype=float)
vals_x2 = np.array([x2.subs(y, i) for i in vals_y], dtype=float)
vals_y_i = np.linspace(min_y, max_y, 500)
vals_x1_i = np.array([x1.subs(y, i) for i in vals_y_i], dtype=float)
vals_x2_i = np.array([x2.subs(y, i) for i in vals_y_i], dtype=float)

# Gráfico
#pyplot.gca().set_aspect('equal', 'datalim')
pyplot.xlabel(r"$x$")
pyplot.ylabel(r"$y$")
pyplot.axhline(y=0, linestyle="--", color="black")
pyplot.axhline(y=min_y, linestyle="--", color="blue")
pyplot.axhline(y=max_y, linestyle="--", color="blue")

pyplot.plot(vals_x1, vals_y, color="red", label=r"$y^2$")
pyplot.plot(vals_x2, vals_y, color="red", label=r"$\sqrt{y}$")
pyplot.fill_betweenx(vals_y_i, vals_x1_i, x2=vals_x2_i, step="pre", alpha=0.4)

# Mostra resultados
area1 = sp.integrate(x1, (y, min_y, max_y))
area2 = sp.integrate(x2, (y, min_y, max_y))
print(f"Solução: ({np.abs(area1 - area2)}) -> ~{np.abs((area1 - area2).evalf())}\n")
pyplot.legend()
pyplot.show()
```
