```python
import matplotlib.pyplot as pyplot
import sympy as sp


t = sp.Symbol("t")
s0 = 800
v0 = 0
a = 9.81


# Funções
v = v0 + a * t
s = s0 + v0 * t - a/2 * t**2

d_v = sp.diff(v, t)
d_s = sp.diff(s, t)


# Parâmetros para t
min_t = 0
max_t = 20
num_points = 40


# Cálculo dos valores de t
vals_t = [i * (max_t - min_t)/num_points for i in range(num_points)]


# Gráfico
#pyplot.gca().set_aspect('equal', 'datalim')
pyplot.plot(vals_t, [s.subs(t, i) for i in vals_t])
pyplot.plot(vals_t, [v.subs(t, i) for i in vals_t])
pyplot.plot(vals_t, [a for _ in vals_t])
pyplot.show()
```
