#Exercício Integrais Definidas


Considerando as integrais a seguir você deverá plotar o
gráfico de cada uma das equações que estão sendo
integradas e marcar a área entre os limites de integração.
Depois, usando apenas álgebra, calcular a área sobre estas
curvas e, finalmente, verificar seus cálculos calculando a área
sobre estas curvas usando a Soma de Reimann. Esta Tarefa
deve ser realizada no Google Colab usando o Python, limitado
ao uso das bibliotecas numpy, sympy e matplotlib.


```python
# Função para o cálculo das integrais e mostrar seu gráfico

import matplotlib.pyplot as pyplot
import numpy as np
import sympy as sp

def plot_int(x, f, min_x:float, max_x:float, x_i_start:float, x_i_stop:float, tryAproximate:bool=False):
  # Calcula a integral de f(x)
  i_f = sp.integrate(f, (x, x_i_start, x_i_stop))

  # Cálculo dos valores de x e de f(x)
  vals_x = np.linspace(min_x, max_x, 500)
  vals_f = np.array([f.subs(x, i) for i in vals_x], dtype=float)

  vals_x_i = np.linspace(x_i_start, x_i_stop, 500)
  vals_f_i = np.array([f.subs(x, i) for i in vals_x_i], dtype=float)

  # Gráfico
  #pyplot.gca().set_aspect('equal', 'datalim')
  pyplot.axhline(y=0, linestyle="--", color="black")
  pyplot.axvline(x=x_i_start, linestyle="--", color="blue")
  pyplot.axvline(x=x_i_stop, linestyle="--", color="blue")

  pyplot.plot(vals_x, vals_f, color="red")
  pyplot.fill_between(vals_x_i, vals_f_i, step="pre", alpha=0.4)

  # Mostra resultados
  if not tryAproximate:
    print(f"Solução: ({i_f})\n")
  else:
    print(f"Solução: ({i_f}) -> ~{i_f.evalf()}\n")
  pyplot.show()
```

a) $\int_{2.5}^{7}(40x^3+12x^2-9x) dx$

```python
x = sp.Symbol("x")
f = 40*x**3 + 12*x**2 - 9*x
plot_int(x=x, f=f, min_x=0, max_x=10, x_i_start=2.5, x_i_stop=7)
```

b) $\int_{7.5}^{10}(12t^7-t^2-t+3)dt$

```python
t = sp.Symbol("t")
f = 12*t**7 + t**2 - t + 3
plot_int(x=t, f=f, min_x=5, max_x=12, x_i_start=7.5, x_i_stop=10)
```

c) $\int_{1}^{2}(\frac{2x^5-x+5}{x^2})dx$

```python
x = sp.Symbol("x")
f = (2*x**2 - x + 5) / (x**2)
plot_int(x=x, f=f, min_x=0.5, max_x=2.5, x_i_start=1, x_i_stop=2, tryAproximate=True)
```

d) $\int_{0}^{\frac{\pi}{2}}(7\sin{x}-2\cos{x})dx$

```python
x = sp.Symbol("x")
f = 7*sp.sin(x) - 2*sp.cos(x)
plot_int(x=x, f=f, min_x=-1, max_x=3, x_i_start=0, x_i_stop=np.pi/2)
```

e) $\int_{-\pi}^{\pi}(5\sin{2x}+3^x)dx$

```python
x = sp.Symbol("x")
f = 5*sp.sin(2*x) + 3**x
plot_int(x=x, f=f, min_x=-4, max_x=4, x_i_start=-np.pi, x_i_stop=np.pi, tryAproximate=True)
```
