```python
import matplotlib.pyplot as pyplot
import numpy as np
import sympy as sp


t = sp.Symbol("t")
t_i_start = 1
t_i_stop = 4


# Funções
f = t**2 * sp.cos(t)
i_f_1_4 = sp.integrate(f, (t, t_i_start, t_i_stop))


# Parâmetros para t
min_t = 0
max_t = 5


# Cálculo dos valores de t e de f(t)
vals_t = np.linspace(min_t, max_t, 500)
vals_t_i = np.linspace(t_i_start, t_i_stop, 500)
vals_f = np.array([f.subs(t, i) for i in vals_t], dtype=float)
vals_f_i = np.array([f.subs(t, i) for i in vals_t_i], dtype=float)


# Gráfico
#pyplot.gca().set_aspect('equal', 'datalim')
pyplot.plot(vals_t, vals_f, color="red")
pyplot.fill_between(vals_t_i, vals_f_i, step="pre", alpha=0.4)
pyplot.axvline(x=t_i_start, linestyle="--", color="blue")
pyplot.axvline(x=t_i_stop, linestyle="--", color="blue")
pyplot.show()
```
