```python
# Importando as bibliotecas
import pyspark  # Biblioteca principal
from pyspark.sql import SparkSession  # Biblioteca para a criação de uma sessão Spark
```

```python
# Criação de uma sessão Spark
spark = SparkSession.builder.appName("Word Count").master('local[*]').getOrCreate()
```

```python
# Criação do SparkContext (conexão entre o Driver e Workers)
sc = spark.sparkContext
```

```python
rdd = sc.textFile("domcasmurro.txt")  # Adionando dados a um novo rdd
```

```python
rdd.collect()  # Coletando dados do rdd
```

```python
rdd.take(5)  # Alternativa mais leve do que o collect (retorna uma quantidade de exemplares dentro do arquivo)
```

```python
palavras = rdd.map(lambda x: x.lower().replace(".", "").split(" "))  # Transformar cada linha (ou seja, cada item do rdd) em um conjunto de palavras
print(palavras.take(5))

# Mais recomendado usar um FlatMap -> map que achata resultados (lista única, não havendo lista de listas)
palavras = rdd.flatMap(lambda x: x.lower().replace(".", "").split(" "))  # Transformar todas as linhas em um conjunto de palavras
print(palavras.take(5))
```

```python
palavras_filtradas = palavras.filter(lambda x: x != "")  # Filtrando as não vazias
print(palavras_filtradas.take(5))
```

```python
contagem = palavras_filtradas.countByValue()  # Contando cada palavra com uma função pré-feita
print(contagem)
```

```python
print(sorted(contagem.items()))  # Imprimindo contagem ordenada
```

```python
palavras_mapeadas = palavras_filtradas.map(lambda w: (w, 1))  # Map usando chave-valor
print(palavras_mapeadas.take(5))
```

```python
contagem_kv = palavras_mapeadas.reduceByKey(lambda x, y: x + y)  # Contagem usando chave-valor
print(contagem_kv.take(5))
```

```python
contagem_kv.saveAsTextFile("wordcount")  # Salvando como arquivo (igual ao Hadoop)
```

```python
rdd_air = sc.textFile("Airports.csv")
rdd_air.take(5)
```

```python
rdd_usa = rdd_air.filter(lambda x: x.split(",")[4] == "USA")
rdd_usa.take(5)
```

```python
rdd_usa.count()
```

```python
rdd_usa.coalesce(1).saveAsTextFile("airports")
```
